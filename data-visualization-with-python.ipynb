```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Drawing A Line Plot

```python
# Given data set as follows
x = [2, 3, 5, 8]
y = [1, 6, 7, 9]
plt.plot(x,y)
plt.show()
```

2. Giving title for X-label and Y-label


```python
x = [2, 3, 5, 8]
y = [1, 6, 7, 9]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

3. Font size changing

```python
x = [2, 3, 5, 8]
y = [1, 6, 7, 9]
plt.plot(x,y)
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.show()
```

4. Change color & width of the line

```python
x = [2, 3, 5, 8]
y = [1, 6, 7, 9]
plt.plot(x,y, color = 'r', linewidth = 4)
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.show()
```

5. Change marker, marker size & Line style

```python
x = [2, 3, 5, 8]
y = [1, 6, 7, 9]
plt.plot(x,y, color = 'r', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', linestyle = '--')
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.show()
```

6. Add legend to the graph and also change location

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)

plt.plot(x,y, color = 'r', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', linestyle = '--')
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.legend(["A simple line plot"], loc=2)
plt.show()
```

7. Create seperate lines for X and Y and change plot style

```python
plt.style.available
```

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)

plt.plot(x, 'r--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 1")
plt.plot(y, 'b--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 2")
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.legend()
plt.show()
```

8. Xlim and Ylim

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)

plt.plot(x, 'r--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 1")
plt.plot(y, 'b--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 2")
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.xlim(0,20)
plt.ylim(0,20)
plt.legend()
plt.show()
```

9. Change X and Y limit using axis method

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)

plt.plot(x, 'r--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 1")
plt.plot(y, 'b--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 2")
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
# xmin, xmax, ymin, ymax
plt.axis([0,20,0,15])
plt.legend()
plt.show()
```

10. Change X and Y limit using Xticks and Yticks

```python
x_ticks = [x for x in range(0,20,2)]
y_ticks = [y for y in range(0,15,3)]
```

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)

plt.plot(x, 'r--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 1")
plt.plot(y, 'b--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 2")
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

11. Use grid method

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)

plt.plot(x, 'r--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 1")
plt.plot(y, 'b--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 2")
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='g', linestyle='--')
plt.show()
```

12. Change figure style

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)
plt.figure(figsize=(10,3),facecolor='lightpink',edgecolor='black')
plt.plot(x, 'r--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 1")
plt.plot(y, 'b--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 2")
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='g', linestyle='--')
plt.show()
```

13. Save graph

```python
x = np.random.randint(low=1,high=10,size = 20)
y = np.random.randint(low=1,high=8,size=20)
plt.figure(figsize=(10,3),facecolor='lightpink',edgecolor='black')
plt.plot(x, 'r--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 1")
plt.plot(y, 'b--<', linewidth = 4, marker ='<', markersize = 10, markerfacecolor = 'b',
         markeredgecolor = 'b', label = "Line 2")
plt.title("Line Plot", fontsize = 20, loc = 'left')
plt.xlabel("X-axis", fontsize = 20, loc = 'center')
plt.ylabel("Y-axis", fontsize = 20, loc = 'center')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='g', linestyle='--')
plt.savefig('lineplot.png')
plt.show()
```
